```python
import requests
```

```python
url = "https://usis.bracu.ac.bd/academia/admissionRequirement/getAvailableSeatStatus"

try:
    response = requests.get(url)
    response.raise_for_status()

    print("Successfully fetched the content")

    print(response.text)
except requests.exceptions.RequestException as e:
    print(f"An error has occured: {e}")
```

```python
#@title Handling Json Response: if the data is in Json format.

import json

try:
    response = requests.get(url)
    data = response.json()
    print(json.dumps(data, indent=4))
except ValueError:
    print("Response is not in json format")
except requests.exceptions.RequestException as e:
    print(f"An error has occured: {e}")
```

```python
# for html type file

from bs4 import BeautifulSoup

url = "https://usis.bracu.ac.bd/academia/admissionRequirement/getAvailableSeatStatus"
response = requests.get(url)

soup = BeautifulSoup(response.text, "html.parser")

table = soup.find("table")
data_l = []
if(table):
    rows = table.find_all("tr")
    for row in rows:
        cells = row.find_all("td")
        data = [cell.text.strip() for cell in cells]
        dict1 = {data[1]:[data[3], data[5], data[6], data[9]]}
        data_l.append(dict1)
else:
    print("No table found in the html file.")
```

```python
import time

try:
    while(True):
        course_names = ["cse260", "cse321", "cse370", "cse331", "cse422"]

        for course_name in course_names:
            result = [item for item in data_l if course_name.upper() in item and int(item[course_name.upper()][-1]) > 0]
            print(*result, sep="\n")
        print(f"\nAfter 30 seconds:\n")

        time.sleep(30)
except KeyboardInterrupt:
    print("The programme has been interrupted.")
```

```python
#@title For Json type file.

import requests

url = "https://usis.bracu.ac.bd/academia/admissionRequirement/getAvailableSeatStatus"
response = requests.get(url)

try:
    # Convert response to JSON
    data = response.json()

    # Pretty print the full JSON response
    import json
    print(json.dumps(data, indent=4))

    # Example: Extracting specific values from JSON data
    # (Modify this according to the actual JSON structure)
    for item in data['seats']:
        print(f"Program: {item['program']}, Available Seats: {item['availableSeats']}")
except ValueError:
    print("The response is not in JSON format.")
except KeyError as e:
    print(f"Key not found: {e}")
```
